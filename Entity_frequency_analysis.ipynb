# CoNLL 2003 Dataset

```python
from datasets import load_dataset
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CoNLL-2003 dataset
dataset = load_dataset("conll2003")

# Combine train, validation, and test sets
data = dataset['train']


training_dataset = dataset['train']
label_list = dataset['train'].features['ner_tags'].feature.names

entities = []
entity_type = []
label_list = dataset['train'].features['ner_tags'].feature.names
# Sample row: {'id': '0', 'tokens': ['EU', 'rejects', 'German', 'call', 'to', 'boycott', 'British', 'lamb', '.'], 'pos_tags': [22, 42, 16, 21, 35, 37, 16, 21, 7], 'chunk_tags': [11, 21, 11, 12, 21, 22, 11, 12, 0], 'ner_tags': [3, 0, 7, 0, 0, 0, 7, 0, 0]}
for item in training_dataset:
    tokens = item['tokens']
    tags = item['ner_tags']
    i = 0
    while i < len(tokens):
      if tags[i] != 0:
        entities.append('')
        entity_type.append(label_list[int(tags[i])])
        while i< len(tokens) and tags[i] != 0:
          entities[-1] = entities[-1] + tokens[i].lower() +  ' '
          i+=1
        entities[-1] = entities[-1].strip()
      else:
        i+=1
# Entity Type count
entity_type_df = pd.DataFrame(entity_type, columns=['Entity'])
entity_type_counts = entity_type_df['Entity'].value_counts()

# Entity count
entities_df = pd.DataFrame(entities, columns=['Entity'])
entity_counts = entities_df['Entity'].value_counts()
entity_counts_df = entity_counts.reset_index()
entity_counts_df.columns = ['Entity', 'Count']



# Plot the distribution of Entity Type count
sns.barplot(x=entity_type_counts.index, y=entity_type_counts.values)
plt.title('Distribution of Entity Types in CoNLL-2003 Dataset')
plt.xlabel('Entity Type')
plt.ylabel('Frequency')
plt.show()


# Plot the distribution of specific entity occurrences
sns.barplot(data=entity_counts_df.head(30), x='Count', y='Entity')
plt.title('Top 30 Entities in CoNLL-2003 Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
#Total entities : 8286
sns.barplot(data=entity_counts_df.tail(30), x='Count', y='Entity')
plt.title('Bottom 30 Entities in CoNLL-2003 Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
```

# Onto Notes 5

```python
from datasets import load_dataset
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

###########################################################################################################################################

# Load the CoNLL-2003 dataset
dataset = load_dataset("ontonotes/conll2012_ontonotesv5", "english_v4")

# Combine train, validation, and test sets
data = dataset['train']


training_dataset = dataset['train']
label_list = dataset['train'].features['sentences'][0]['named_entities'].feature.names

entities = []
entity_type = []
for item in training_dataset:
    for sentence in item['sentences']:
      tokens = sentence['words']
      tags = sentence['named_entities']
      i = 0
      while i < len(tokens):
        if tags[i] != 0:
          entities.append('')
          entity_type.append(label_list[int(tags[i])])
          while i< len(tokens) and tags[i] != 0:
            entities[-1] = entities[-1] + tokens[i].lower() +  ' '
            i+=1
          entities[-1] = entities[-1].strip()
        else:
          i+=1
# Entity Type count
entity_type_df = pd.DataFrame(entity_type, columns=['Entity'])
entity_type_counts = entity_type_df['Entity'].value_counts()

# Entity count
entities_df = pd.DataFrame(entities, columns=['Entity'])
entity_counts = entities_df['Entity'].value_counts()
entity_counts_df = entity_counts.reset_index()
entity_counts_df.columns = ['Entity', 'Count']



# Plot the distribution of Entity Type count
sns.barplot(x=entity_type_counts.index, y=entity_type_counts.values)
plt.title('Distribution of Entity Types in Ontonotes5 Dataset')
plt.xlabel('Entity Type')
plt.ylabel('Frequency')
plt.show()
print(entity_type_counts)

# Plot the distribution of specific entity occurrences
sns.barplot(data=entity_counts_df.head(30), x='Count', y='Entity')
plt.title('Top 30 Entities in Ontonotes5 Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
#Total entities : 25471
sns.barplot(data=entity_counts_df.tail(30), x='Count', y='Entity')
plt.title('Bottom 30 Entities in Ontonotes5 Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
```

## FEW-NERD Dataset for Fine grained entity typing

```python
from datasets import load_dataset
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CoNLL-2003 dataset
dataset = load_dataset("DFKI-SLT/few-nerd", "supervised")

# Combine train, validation, and test sets
data = dataset['train']


training_dataset = dataset['train']
label_list_coarse = dataset['train'].features['ner_tags'].feature.names
label_list_fine = dataset['train'].features['fine_ner_tags'].feature.names

entities = []
entity_type = []
fine_entity_type = []
# Sample rne', 'tokens': ['EU', 'rejects', 'German', 'call', 'to', 'boycott', 'British', 'lamb', '.'], 'pos_tags': [22, 42, 16, 21, 35, 37, 16, 21, 7], 'chunk_tags': [11, 21, 11, 12, 21, 22, 11, 12, 0], 'ner_tags': [3, 0, 7, 0, 0, 0, 7, 0, 0]}
for item in training_dataset:
      tokens = item['tokens']
      tags = item['ner_tags']
      fine_tags = item['fine_ner_tags']
      i = 0
      while i < len(tokens):
        if tags[i] != 0:
          entities.append('')
          entity_type.append(label_list_coarse[int(tags[i])])
          fine_entity_type.append(label_list_fine[int(fine_tags[i])])
          while i< len(tokens) and tags[i] != 0:
            entities[-1] = entities[-1] + tokens[i].lower() +  ' '
            i+=1
          entities[-1] = entities[-1].strip()
        else:
          i+=1
# Entity Type count
entity_type_df = pd.DataFrame(entity_type, columns=['Entity'])
entity_type_counts = entity_type_df['Entity'].value_counts()

# Fine Entity Type count
fine_entity_type_df = pd.DataFrame(fine_entity_type, columns=['Fine Entity'])
fine_entity_type_counts = fine_entity_type_df['Fine Entity'].value_counts()

# Entity count
entities_df = pd.DataFrame(entities, columns=['Entity'])
entity_counts = entities_df['Entity'].value_counts()
entity_counts_df = entity_counts.reset_index()
entity_counts_df.columns = ['Entity', 'Count']


# Plot the distribution of Entity Type count
sns.barplot(x=entity_type_counts.index, y=entity_type_counts.values)
plt.title('Distribution of Entity Types in FEW NERD Dataset')
plt.xlabel('Entity Type')
plt.ylabel('Frequency')
plt.show()
print(entity_type_counts)

# Plot the distribution of Entity Type count
sns.barplot(x=fine_entity_type_counts.index, y=fine_entity_type_counts.values)
plt.title('Distribution of Entity Types in FEW NERD Dataset')
plt.xlabel('Fine Entity Type')
plt.ylabel('Frequency')
plt.show()
print(fine_entity_type_counts)

# Plot the distribution of specific entity occurrences
sns.barplot(data=entity_counts_df.head(30), x='Count', y='Entity')
plt.title('Top 30 Entities in FEW NERD Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
#Total entities : 183167
sns.barplot(data=entity_counts_df.tail(30), x='Count', y='Entity')
plt.title('Bottom 30 Entities in FEW NERD Dataset')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
entity_counts_df
```
